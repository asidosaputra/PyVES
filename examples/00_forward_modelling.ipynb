## Import library

```
import numpy as np
import matplotlib.pyplot as plt

from forward import SLB
```

### Load data observation
In this case, we use data from samples data


```
fname = './samples/sample1.txt'
data = np.loadtxt(fname)    #data
ab2 = data[:, 0]            #half space of electrode
rhoap_obs = data[:, 1]      #apperant resistivities from observation
```

## Initial Model
Model initial are the resistivities and thickness of layers

```
rhotr = np.array([80, 60, 30])  #initial apperant resistivities in ohm.m
thick = np.array([5, 10])       #intial thickness in m.
```

## Forward Modelling

```
lf = SLB()      #make new object of SLB() Class.
lf.run(ab2, rhoap_obs, rhotr, thick, filter_coeff='guptasarma_7') #run forward with specify filter coefficient
lf.plot_mod(save_fig = True)    #ploting 
```
